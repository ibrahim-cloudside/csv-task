Filling missing value as Unknown

```python
nt['director'].fillna('Unknown', inplace=True)
nt['cast'].fillna('Unknown', inplace=True)
nt['country'].fillna('Unknown', inplace=True)
nt['date_added'].fillna('Unknown', inplace=True)
nt['duration'].fillna('Unknown', inplace=True)
nt['rating'].fillna('Unknown', inplace=True)
```

```python
nt.isnull().sum()
```

```python
movies = nt[nt['type'] == 'Movie']
print("Total count of movie: ",len(movies))
print('***************************')
movies.head()
```

```python
tv_show = nt[nt['type'] == 'TV Show']
print("Total count of TV Show: ",len(tv_show))
print('******************************')
tv_show.head()
```
